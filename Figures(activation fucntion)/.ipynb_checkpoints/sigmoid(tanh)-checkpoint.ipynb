```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42

custom_params = {"axes.spines.right": False, "axes.spines.top": False,
                 "font.family"   : "Times New Roman", "font.scale": 1.5}

sns.set_style(style="ticks", rc=custom_params) #设置绘图风格
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    """
    Sigmoid activation function.
    """
    return 1 / (1 + np.exp(-x))

# Define the tanh function
def tanh(x):
    """
    Hyperbolic tangent activation function.
    """
    return np.tanh(x)

# Define the sigmoid(tanh(x)) function
def sigmoid_of_tanh(x):
    """
    Composite function: sigmoid applied to tanh(x).
    """
    return sigmoid(tanh(x))

# Define the derivative of the sigmoid function
def derivative_sigmoid(x):
    """
    Derivative of the sigmoid function.
    """
    s = sigmoid(x)
    return s * (1 - s)

# Define the derivative of the tanh function
def derivative_tanh(x):
    """
    Derivative of the tanh function.
    """
    return 1 - tanh(x)**2

# Define the derivative of sigmoid(tanh(x))
def derivative_sigmoid_of_tanh(x):
    """
    Derivative of the composite sigmoid(tanh(x)) function.
    """
    return derivative_sigmoid(tanh(x)) * derivative_tanh(x)

# Generate x values from -5 to 5
x = np.linspace(-5, 5, 100)

# Create a figure with two subplots
plt.figure(figsize=(12, 4))

# First subplot: activation functions
plt.subplot(1, 2, 1)
plt.plot(x, sigmoid(x), label='Sigmoid', color='blue')
plt.plot(x, tanh(x), label='Tanh', color='orange')
plt.plot(x, sigmoid_of_tanh(x), label='Sigmoid(Tanh)', color='red')
plt.xlabel('$x$', fontsize=15)
plt.ylabel('$f(x)$', fontsize=15)
plt.legend()

# Second subplot: derivatives of activation functions
plt.subplot(1, 2, 2)
plt.plot(x, derivative_sigmoid(x), label='Derivative of Sigmoid', color='blue')
plt.plot(x, derivative_tanh(x), label='Derivative of Tanh', color='orange')
plt.plot(x, derivative_sigmoid_of_tanh(x), label='Derivative of Sigmoid(Tanh)', color='red')
plt.xlabel('$x$', fontsize=15)
plt.ylabel('$df(x)/dx$', fontsize=15)
plt.legend(fontsize=10)

# Adjust layout and save the figure
plt.tight_layout()
plt.savefig('./activation_function.pdf', dpi=300, bbox_inches='tight', format='pdf')
plt.show()
```
