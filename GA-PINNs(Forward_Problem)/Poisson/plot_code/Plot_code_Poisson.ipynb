```python
test_loss_1 = np.array(test_loss_1) 
test_loss_2 = np.array(test_loss_2) 
test_loss_3 = np.array(test_loss_3) 
test_loss_4 = np.array(test_loss_4) 
test_loss_5 = np.array(test_loss_5) 
loss_all_1 = np.array(loss_all_1) 
loss_all_2 = np.array(loss_all_2) 
loss_all_3 = np.array(loss_all_3) 
loss_all_4 = np.array(loss_all_4) 
loss_all_5 = np.array(loss_all_5) 
```

```python
times = 800

index1 = np.arange(0,len(test_loss_1),times).astype('int')
index1 = np.append(index1,len(test_loss_1)-1).astype('int')
index2 = np.arange(0,len(test_loss_2),times).astype('int')
index2 = np.append(index2,len(test_loss_2)-1).astype('int')
index3 = np.arange(0,len(test_loss_3),times).astype('int')
index3 = np.append(index3,len(test_loss_3)-1).astype('int')
index4 = np.arange(0,len(test_loss_4),times).astype('int')
index4 = np.append(index4,len(test_loss_4)-1).astype('int')
index5 = np.arange(0,len(test_loss_5),times).astype('int')
index5 = np.append(index5,len(test_loss_5)-1).astype('int')
```

```python
plt.figure(figsize=(8,4))

sns.lineplot(x=np.array(x1)[index1],y=np.array(test_loss_1)[index1],label = 'PINN',c='g', marker="d", markersize=7, markeredgecolor="white", markeredgewidth=1.2)
sns.lineplot(x=np.array(x2)[index2],y=np.array(test_loss_2)[index2],label = 'GA-PINN',c='b', marker="s", markersize=7, markeredgecolor="white", markeredgewidth=1.2)
sns.lineplot(x=np.array(x3)[index3],y=np.array(test_loss_3)[index3],label = 'PINN$^{\dag}$',c='orange', marker="^", markersize=7, markeredgecolor="white", markeredgewidth=1.2)
sns.lineplot(x=np.array(x4)[index4],y=np.array(test_loss_4)[index4],label = 'GA-PINN$ ^{\dag}$',c='r', marker="o", markersize=7, markeredgecolor="white", markeredgewidth=1.2)
sns.lineplot(x=np.array(x5)[index5],y=np.array(test_loss_5)[index5],label = 'GA-PINN$^{\ddag }$', marker=">", markersize=7, markeredgecolor="white", markeredgewidth=1.2)

plt.xlabel('Iteration',fontsize=20)
plt.ylabel('NRMSE',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)


plt.legend(loc='upper right', frameon=False, fontsize=18, ncol=2)

for spine in plt.gca().spines.values():
    spine.set_edgecolor("#CCCCCC")
    spine.set_linewidth(1.5)
    
plt.yscale('log')

plt.savefig('../figures/Poisson_NRMSE.pdf', dpi=300, bbox_inches='tight', format='pdf')
plt.show()
```
