```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
```

```python
custom_params = {"axes.spines.right": False, "axes.spines.top": False,
                 "font.family"   : "Times New Roman", "font.scale": 1.5}

sns.set_style(style="ticks", rc=custom_params)
```

```python
test_loss_1 = np.load('../experimental_data/method_1/test_loss_1.npy')
test_loss_2 = np.load('../experimental_data/method_2/test_loss_2.npy')
test_loss_3 = np.load('../experimental_data/method_3/test_loss_3.npy')
test_loss_4 = np.load('../experimental_data/method_4/test_loss_4.npy')
test_loss_5 = np.load('../experimental_data/method_5/test_loss_5.npy')

loss_all_1 = np.load('../experimental_data/method_1/loss_all_1.npy')
loss_all_2 = np.load('../experimental_data/method_2/loss_all_2.npy')
loss_all_3 = np.load('../experimental_data/method_3/loss_all_3.npy')
loss_all_4 = np.load('../experimental_data/method_4/loss_all_4.npy')
loss_all_5 = np.load('../experimental_data/method_5/loss_all_5.npy')
```

```python
print('e1:',test_loss_1[-1],'e2:',test_loss_2[-1],'e3:',test_loss_3[-1],'e4:',test_loss_4[-1],'e5:',test_loss_5[-1])
```

```python
print('epochs_1:',len(test_loss_1),'epochs_2:',len(test_loss_2),'epochs_3:',len(test_loss_3),'epochs_4:',len(test_loss_4),'epochs_5:',len(test_loss_5))
```

```python
test_loss_1 = np.array(test_loss_1) 
test_loss_2 = np.array(test_loss_2) 
test_loss_3 = np.array(test_loss_3) 
test_loss_4 = np.array(test_loss_4) 
test_loss_5 = np.array(test_loss_5) 
loss_all_1 = np.array(loss_all_1) 
loss_all_2 = np.array(loss_all_2) 
loss_all_3 = np.array(loss_all_3) 
loss_all_4 = np.array(loss_all_4) 
loss_all_5 = np.array(loss_all_5) 
```

```python
x1 = np.arange(len(test_loss_1))+1
x2 = np.arange(len(test_loss_2))+1
x3 = np.arange(len(test_loss_3))+1
x4 = np.arange(len(test_loss_4))+1
x5 = np.arange(len(test_loss_5))+1
times = 2602
x_max = 15000
x1_index = np.arange(0,len(test_loss_1),times)
x2_index = np.arange(0,len(test_loss_2),times)
x3_index = np.arange(0,len(test_loss_3),times)
x4_index = np.arange(0,len(test_loss_4),times)
x5_index = np.arange(0,len(test_loss_5),times)

x1_index = np.concatenate((x1_index,[len(test_loss_1)-1]),axis=0)
x2_index = np.concatenate((x2_index,[len(test_loss_2)-1]),axis=0)
x3_index = np.concatenate((x3_index,[len(test_loss_3)-1]),axis=0)
x4_index = np.concatenate((x4_index,[len(test_loss_4)-1]),axis=0)
x5_index = np.concatenate((x5_index,[len(test_loss_5)-1]),axis=0)
```

```python
plt.figure(figsize=(8,4))

sns.lineplot(x=x1_index+1,y=test_loss_1[x1_index],label = 'PINNs',c='g', marker="d", markersize=7, markeredgecolor="white", markeredgewidth=1.2)
sns.lineplot(x=x2_index+1,y=test_loss_2[x2_index],label = 'GA-PINNs',c='b', marker="s", markersize=7, markeredgecolor="white", markeredgewidth=1.2)
sns.lineplot(x=x3_index+1,y=test_loss_3[x3_index],label = 'PINNs$^{\dag}$',c='orange', marker="^", markersize=7, markeredgecolor="white", markeredgewidth=1.2)
sns.lineplot(x=x4_index+1,y=test_loss_4[x4_index],label = 'GA-PINNs$ ^{\dag}$',c='r', marker="o", markersize=7, markeredgecolor="white", markeredgewidth=1.2)
sns.lineplot(x=x5_index+1,y=test_loss_5[x5_index],label = 'GA-PINNs$^{\ddag }$', marker=">", markersize=7, markeredgecolor="white", markeredgewidth=1.2)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.legend(loc='upper right', frameon=False, fontsize=18, ncol=2)

for spine in plt.gca().spines.values():
    spine.set_edgecolor("#CCCCCC")
    spine.set_linewidth(1.5)
    
plt.yscale('log') 

plt.savefig('../figures/Burgers_NRMSE.pdf', dpi=300, bbox_inches='tight', format='pdf')
plt.show()
```

```python
plt.figure(figsize=(8,4))

loss_all_1 = np.array(loss_all_1)
loss_all_2 = np.array(loss_all_2)
loss_all_3 = np.array(loss_all_3)
loss_all_4 = np.array(loss_all_4)
loss_all_5 = np.array(loss_all_5)

sns.lineplot(x=x1_index+1,y=loss_all_1[x1_index],label = 'PINNs',c='g', marker="d", markersize=7, markeredgecolor="white", markeredgewidth=1.2)
sns.lineplot(x=x2_index+1,y=loss_all_2[x2_index],label = 'GA-PINNs',c='b', marker="s", markersize=7, markeredgecolor="white", markeredgewidth=1.2)
sns.lineplot(x=x3_index+1,y=loss_all_3[x3_index],label = 'PINNs$^{\dag}$',c='orange', marker="^", markersize=7, markeredgecolor="white", markeredgewidth=1.2)
sns.lineplot(x=x4_index+1,y=loss_all_4[x4_index],label = 'GA-PINNs$ ^{\dag}$',c='r', marker="o", markersize=7, markeredgecolor="white", markeredgewidth=1.2)
sns.lineplot(x=x5_index+1,y=loss_all_5[x5_index],label = 'GA-PINNs$^{\ddag }$', marker=">", markersize=7, markeredgecolor="white", markeredgewidth=1.2)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.legend(loc='upper right', frameon=False, fontsize=18, ncol=2)

for spine in plt.gca().spines.values():
    spine.set_edgecolor("#CCCCCC")
    spine.set_linewidth(1.5)
    
plt.yscale('log')

plt.savefig('../figures/Burgers_loss_pi.pdf', dpi=300, bbox_inches='tight', format='pdf')
plt.show()
```
