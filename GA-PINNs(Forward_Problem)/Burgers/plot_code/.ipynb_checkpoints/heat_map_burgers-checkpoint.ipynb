```python
import numpy as np
import torch
from torch import nn, optim, autograd
from torch.nn import functional as F
from pyDOE import lhs
import scipy.io
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.gridspec as gridspec
%matplotlib inline
from mpl_toolkits.axes_grid1 import make_axes_locatable

from models_all import *

#Paper reproduction
torch.manual_seed(1234)
torch.cuda.manual_seed(1234)
np.random.seed(1234)
```

```python
import scipy.io
data = scipy.io.loadmat('./burgers_shock.mat')
t_ = data['t'].flatten()[:,None]
x_ = data['x'].flatten()[:,None]
Exact = np.real(data['usol']).T

X, T = np.meshgrid(x_,t_)

X_star = np.hstack((T.flatten()[:,None], X.flatten()[:,None]))
u_star = Exact.flatten()[:,None]  
```

```python
PINNs1 = torch.load('../saved_model/PINNs1')
PINNs2 = torch.load('../saved_model/PINNs2')
PINNs3 = torch.load('../saved_model/PINNs3')
PINNs4 = torch.load('../saved_model/PINNs4')
PINNs5 = torch.load('../saved_model/PINNs5')
```

```python
X_star = torch.from_numpy(X_star).float()
```

```python
e1 = np.abs(PINNs1(X_star).detach().numpy()-u_star)
e2 = np.abs(PINNs2(X_star).detach().numpy()-u_star)
e3 = np.abs(PINNs3(X_star).detach().numpy()-u_star)
e4 = np.abs(PINNs4(X_star).detach().numpy()-u_star)
e5 = np.abs(PINNs5(X_star).detach().numpy()-u_star)
```

```python
vmin=np.min(e2)
vmax=np.max(e2)
```

```python
fig, ax = plt.subplots()
gs0 = gridspec.GridSpec(1, 2)
gs0.update(top=1-0.06, bottom=1-1/3, left=0.15, right=0.85, wspace=0)
ax = plt.subplot(gs0[:, :])

plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

h = ax.imshow(e1.reshape(Exact.shape).T, interpolation='nearest', cmap='rainbow',  extent=[t_.min(), t_.max(), x_.min(), x_.max()],
            origin='lower', aspect='auto',vmin=vmin,
vmax=vmax)
divider = make_axes_locatable(ax)

plt.savefig('../figures/Burgers_abs_loss_1.pdf',format='pdf',dpi=300, bbox_inches = 'tight')
plt.show()
```
