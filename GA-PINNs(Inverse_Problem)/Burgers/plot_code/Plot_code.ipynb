```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
```

```python
custom_params = {"axes.spines.right": False, "axes.spines.top": False,
                 "font.family"   : "Times New Roman", "font.scale": 20}

sns.set_style(style="ticks", rc=custom_params)
```

# Test Loss

```python
J1_test_loss_1 = np.load('../experimental_data/J1/test_loss_1.npy')
J1_test_loss_2 = np.load('../experimental_data/J1/test_loss_2.npy')
```

```python
plt.figure(figsize=(8,4))

# iterations
x = [1000,2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000]

sns.lineplot(x=x, y=J1_test_loss_1, color='b', linewidth=2.0, marker="o", markersize=8, markeredgecolor="white", markeredgewidth=1.2, label=r'PINN$^{\S}$')
sns.lineplot(x=x, y=J1_test_loss_2, color='r', linewidth=2.0, marker="s", markersize=8, markeredgecolor="white", markeredgewidth=1.2, label=r'GA-PINN$^{\S}$')

plt.legend(loc='upper right', frameon=False, fontsize=20)

plt.xticks(fontsize=15)
plt.yticks(fontsize=20)

for spine in plt.gca().spines.values():
    spine.set_edgecolor("#CCCCCC")
    spine.set_linewidth(1.5)
    
plt.yscale('log') 

plt.savefig('../figures/J1_test_loss.pdf', dpi=300, bbox_inches='tight', format='pdf')

plt.show()
```

# undetermined parameter

```python
# iterations
x = np.array([1,1000,2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000])

J1_C1_list = np.load('../experimental_data/J1/C1_list.npy')[x-1]
J1_C2_list = np.load('../experimental_data/J1/C2_list.npy')[x-1]
```

```python
plt.figure(figsize=(8,4))

sns.lineplot(x=x, y=J1_C1_list, color='b', linewidth=2.0, marker="o", markersize=8, markeredgecolor="white", markeredgewidth=1.2, label=r'PINN$^{\S}$')
sns.lineplot(x=x, y=J1_C2_list, color='r', linewidth=2.0, marker="s", markersize=8, markeredgecolor="white", markeredgewidth=1.2, label=r'GA-PINN$^{\S}$')

plt.axhline(y=0.01/np.pi, color='green', linestyle='--', linewidth=2, label=r'v = $\frac{0.01}{\pi}$')

plt.legend(loc='upper right', frameon=False, fontsize=20)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

for spine in plt.gca().spines.values():
    spine.set_edgecolor("#CCCCCC")
    spine.set_linewidth(1.5)
    
plt.savefig('../figures/J1_C_test.pdf', dpi=300, bbox_inches='tight', format='pdf')
plt.show()
```
