```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
```

```python
custom_params = {"axes.spines.right": False, "axes.spines.top": False,
                 "font.family"   : "Times New Roman", "font.scale": 20}

sns.set_style(style="ticks", rc=custom_params)
```

# Test Loss

```python
J1_test_loss_1 = np.load('../experimental_data/J1/test_loss_1.npy')
J1_test_loss_2 = np.load('../experimental_data/J1/test_loss_2.npy')
```

```python
plt.figure(figsize=(8,4))

# iterations
x = [500, 1000, 1500, 2000, 2500, 3000 , 3500, 4000, 4500, 5000, 5500, 6000, 6500, 7000]

sns.lineplot(x=x, y=J1_test_loss_1, color='b', marker="o", markersize=8, markeredgecolor="white", markeredgewidth=1.2, label=r'PINN$^\S$')
sns.lineplot(x=x, y=J1_test_loss_2, color='r', marker="s", markersize=8, markeredgecolor="white", markeredgewidth=1.2, label=r'GA-PINN$^\S$')

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.legend(loc='upper right', frameon=False, fontsize=20)

for spine in plt.gca().spines.values():
    spine.set_edgecolor("#CCCCCC")
    spine.set_linewidth(1.5)
    
plt.yscale('log') 

plt.savefig('../figures/J1_test_loss.pdf', dpi=300, bbox_inches='tight', format='pdf')
plt.show()
```

# undetermined parameter

```python
J1_C1_list = np.load('../experimental_data/J1/C1_list.npy')
J1_C2_list = np.load('../experimental_data/J1/C2_list.npy')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 4))

# iterations
x = [500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500, 6000, 6500, 7000]

scaled_J1_C1_list = [value * 1000 for value in J1_C1_list]
scaled_J1_C2_list = [value * 1000 for value in J1_C2_list]

sns.lineplot(x=x, y=scaled_J1_C1_list, color='b', linewidth=2.0, marker="o", markersize=8, markeredgecolor="white", markeredgewidth=1.2, label=r'PINN$^\S$')
sns.lineplot(x=x, y=scaled_J1_C2_list, color='r', linewidth=2.0, marker="s", markersize=8, markeredgecolor="white", markeredgewidth=1.2, label=r'GA-PINN$^\S$')

plt.axhline(y=1, color='green', linestyle='--', linewidth=2, label='D = 0.001')

plt.legend(loc='upper right', frameon=False, fontsize=20)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(1-0.2, 4.5)

plt.annotate(r'$\times 10^{-3}$', xy=(-0.05, 1.14), xycoords='axes fraction', fontsize=15, verticalalignment='top')

for spine in plt.gca().spines.values():
    spine.set_edgecolor("#CCCCCC")
    spine.set_linewidth(1.5)

plt.savefig('../figures/J1_C_test.pdf', dpi=300, bbox_inches='tight', format='pdf')
plt.show()
```
