```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
```

```python
custom_params = {"axes.spines.right": False, "axes.spines.top": False,
                 "font.family"   : "Times New Roman", "font.scale": 20}

sns.set_style(style="ticks", rc=custom_params)
```

### $\sigma$

```python
plt.figure(figsize=(8,4))

x = np.arange(1001,21001,1000)-1
y1 = np.load('../experimental_data/J2/C1_a_list.npy')[1:]
y2 = np.load('../experimental_data/J2/C1_b_list.npy')[1:]

sns.lineplot(x=x, y=y1, color='b', linewidth=2.0, marker="o", markersize=6, markeredgecolor="white", markeredgewidth=1.2, label=r'PINN$^{\S}$')
sns.lineplot(x=x, y=y2, color='r', linewidth=2.0, marker="s", markersize=6, markeredgecolor="white", markeredgewidth=1.2, label=r'GA-PINN$^{\S}$')

plt.axhline(y=10, color='green', linestyle='--', linewidth=2, label=r'$\sigma = 10$')

plt.legend(loc='lower right', frameon=False, fontsize=20)

plt.xticks(fontsize=15)
plt.yticks(fontsize=20)

for spine in plt.gca().spines.values():
    spine.set_edgecolor("#CCCCCC")
    spine.set_linewidth(1.5)

plt.savefig('../figures/J2_C1.pdf', dpi=300, bbox_inches='tight', format='pdf')
plt.show()
```
