```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
%matplotlib inline
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
```

```python
custom_params = {"axes.spines.right": False, "axes.spines.top": False,
                 "font.family"   : "Times New Roman", "font.scale": 1.5}

sns.set_style(style="ticks", rc=custom_params)
```

```python
color1 = "#038355" 
color2 = "#ffc34e" 
```

```python
folder_path = '../experimental_data/PINN/'

loss_T_1 = np.load(os.path.join(folder_path, 'loss_T_1.npy'))
test_loss_1 = np.load(os.path.join(folder_path, 'test_loss_1.npy'))
C1_list = np.load(os.path.join(folder_path, 'C1_list.npy'))
```

```python
folder_path = '../experimental_data/GA_PINN/'

GA_loss_T_1 = np.load(os.path.join(folder_path, 'loss_T_1.npy'))
GA_test_loss_1 = np.load(os.path.join(folder_path, 'test_loss_1.npy'))
GA_C1_list = np.load(os.path.join(folder_path, 'C1_list.npy'))
```

```python
fig, ax = plt.subplots(figsize=(8,5))

fig.patch.set_facecolor('white')


c = (234,234,242)
matplotlib_color = tuple(value/255 for value in c)

plt.plot(np.arange(len(test_loss_1)),test_loss_1, c=color1, linewidth=1.2,label='PINN')
plt.plot(np.arange(len(GA_test_loss_1)),GA_test_loss_1, c=color2, linewidth=1.2,label='GA-PINN')

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.legend(loc='upper right', fontsize=25)

plt.yscale('log')
plt.xlabel('Iterations', fontsize=25)
plt.ylabel('NRMSE', fontsize=25)

ax.grid(True, color='lightgray', linestyle='-', linewidth=0.5)

plt.savefig('../figures/Burgers_NRMSE.pdf', dpi=66, bbox_inches='tight', format='pdf')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,5))

fig.patch.set_facecolor('white')


c = (234,234,242)
matplotlib_color = tuple(value/255 for value in c)

plt.axhline(y=0.01/np.pi, color='k', linestyle='--', linewidth=1.5, label=r'$\theta^*=0.01/\pi$',alpha=0.9)

plt.plot(np.arange(len(C1_list)),C1_list, c=color1, linewidth=1.2,label='PINN')
plt.plot(np.arange(len(GA_C1_list)),GA_C1_list, c=color2, linewidth=1.2,label='GA-PINN')

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.legend(loc='upper right', fontsize=25)

plt.yscale('log')
plt.xlabel('Iterations', fontsize=25)
plt.ylabel('Estimated parameter', fontsize=25)

ax.grid(True, color='lightgray', linestyle='-', linewidth=0.5)

plt.savefig('../figures/Burgers_C.pdf', dpi=66, bbox_inches='tight', format='pdf')
plt.show()
```
